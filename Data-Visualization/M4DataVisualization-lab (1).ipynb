# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt

QUERY = """
SELECT ConvertedComp
FROM master
"""

df = pd.read_sql_query(QUERY, conn)

# Assuming df is your DataFrame and 'ConvertedComp' is the column you want to plot
plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'].dropna(), bins=30, color='blue', edgecolor='black')
plt.title('Distribution of ConvertedComp')
plt.xlabel('ConvertedComp')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Age
FROM master
"""

# Execute the SQL query and load the data into a pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

plt.figure(figsize=(8, 6))
plt.boxplot(df['Age'].dropna(), vert=True, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Box Plot of Age')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""

# Execute the SQL query and load the data into a pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

df_filtered = df.dropna(subset=['Age', 'WorkWeekHrs'])

plt.figure(figsize=(10, 6))
plt.scatter(df_filtered['Age'], df_filtered['WorkWeekHrs'], alpha=0.5, color='blue')
plt.title('Scatter Plot of Age vs WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""

# Execute the SQL query and load the data into a pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

df_filtered = df.dropna(subset=['WorkWeekHrs', 'CodeRevHrs', 'Age'])

# Plotting the bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(df_filtered['WorkWeekHrs'], df_filtered['CodeRevHrs'], 
            s=df_filtered['Age'], alpha=0.5, color='blue')

plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs (Bubble Size: Age)')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

QUERY = """
SELECT DatabaseDesireNextYear, COUNT(*) as count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
LIMIT 5
"""

df = pd.read_sql_query(QUERY, conn)

print(df)

labels = df['DatabaseDesireNextYear']
sizes = df['count']

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=plt.cm.Paired(range(len(labels))))
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""

df = pd.read_sql_query(QUERY, conn)

df_filtered = df.dropna(subset=['WorkWeekHrs', 'CodeRevHrs'])

# Calculate median WorkWeekHrs and CodeRevHrs for the age group
median_values = df_filtered.groupby('Age').agg({
    'WorkWeekHrs': 'median',
    'CodeRevHrs': 'median'
}).reset_index()

# Display the median values to verify
print(median_values)

median_values.set_index('Age', inplace=True)

# Plotting the stacked bar chart
plt.figure(figsize=(10, 6))

# Plot WorkWeekHrs and CodeRevHrs as stacked bars
plt.bar(median_values.index, median_values['WorkWeekHrs'], label='WorkWeekHrs', color='lightblue')
plt.bar(median_values.index, median_values['CodeRevHrs'], bottom=median_values['WorkWeekHrs'], label='CodeRevHrs', color='lightcoral')

plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.legend()
plt.grid(True)

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
"""

# Execute the SQL query and load the data into a pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

df_filtered = df.dropna(subset=['ConvertedComp'])

# Calculate median ConvertedComp for each age
median_values = df_filtered.groupby('Age').agg({
    'ConvertedComp': 'median'
}).reset_index()

# Display the median values to verify
print(median_values)
plt.figure(figsize=(10, 6))
plt.plot(median_values['Age'], median_values['ConvertedComp'], marker='o', linestyle='-', color='b')

# Adding labels and title
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.title('Median ConvertedComp by Age (45 to 60)')
plt.grid(True)

# Show the plot
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

QUERY = """
SELECT MainBranch
FROM master
"""

# Execute the SQL query and load the data into a pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

df_filtered = df.dropna(subset=['MainBranch'])

# Count occurrences of each value in MainBranch
counts = df_filtered['MainBranch'].value_counts()

# Convert to DataFrame for easier plotting
counts_df = counts.reset_index()
counts_df.columns = ['MainBranch', 'Count']

# Display the counts DataFrame to verify
print(counts_df)

plt.figure(figsize=(10, 6))
plt.barh(counts_df['MainBranch'], counts_df['Count'], color='skyblue')

# Adding labels and title
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.title('Distribution of MainBranch')
plt.grid(axis='x')

# Show the plot
plt.show()
```
